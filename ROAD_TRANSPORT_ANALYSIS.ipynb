# **ROAD TRANSPORT ANALYSIS**

The data is gotten from the National Bureau of Statistics website. For easly analysis, we will be merging the data collected for the 4 quarters in 2023 into one.

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/Crashes Q1_ - Sheet1.csv")
df2 = pd.read_csv("/content/Crashes Q2 - Sheet4.csv")
df3 = pd.read_csv("/content/Crashes Q3 - Sheet5.csv")
df4 = pd.read_csv("/content/Crashes Q4 - Sheet5.csv")
```

```python
#concating/joining our data
data = pd.concat([df1, df2, df3, df4])
```
